### Clean `emp_title`

Look at top 20 titles

```python
loanstats['emp_title'].head(20)

loanstats['emp_title'].value_counts().head(20)
```

How often is `emp_title` null?

```python
loanstats['emp_title'].isnull()

loanstats['emp_title'].isnull().sum()
```

Clean the title and handle missing values

1.   Capitalize
2.  Strip spaces
3. Replace NaN with 'Missing'


```python
import numpy as np 

examples = ['owner', 'Supervisor ', ' Project Manager', np.nan]

def clean_title(x): 
  if isinstance(x, str): 
   return x.strip().title()
  else: 
    return 'Unknown'
  
[clean_title(x) for x in examples]


      
```

```python
loanstats['emp_title'].head(10)
```

```python
loanstats['emp_title'].apply(clean_title).head(10)

```

## Work with dates

pandas documentation
- [to_datetime](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_datetime.html)
- [Time/Date Components](https://pandas.pydata.org/pandas-docs/stable/timeseries.html#time-date-components) "You can access these properties via the `.dt` accessor"

```python
loanstats['issue_d'].head().values

```

```python
loanstats['issue_d'].describe()
```

```python
loanstats['issue_d'] = pd.to_datetime(loanstats['issue_d'], infer_datetime_format=True)

```

```python
loanstats['issue_d'].head().values
```

```python
loanstats['issue_d'].describe()
```

```python
loanstats['issue_year'] = loanstats['issue_d'].dt.year

loanstats['issue_year']
```

```python
loanstats['issue_month'] = loanstats['issue_d'].dt.month

loanstats['issue_month'] 
```

```python
pd.to_datetime(loanstats['earliest_cr_line'], infer_datetime_format=True)

loanstats['earliest_cr_line'] = pd.to_datetime(loanstats['earliest_cr_line'], infer_datetime_format=True)

loanstats['earliest_cr_line']

```

```python
loanstats['issue_d'] - loanstats['earliest_cr_line']
```

```python
loanstats['days_credit_to_issue'] = (loanstats['issue_d'] - loanstats['earliest_cr_line']).dt.days

loanstats.shape

loanstats['days_credit_to_issue']
```
